```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.stats import norm
```

```python
fig1,ax1 =plt.subplots()
fig2,ax2 = plt.subplots()


x1 = np.linspace(0,1,1000)
y1 = norm.pdf(x1, 0.3, 0.105)
x2 = np.linspace(0,1,1000)
y2 = norm.pdf(x2, 0.7, 0.105)

ax1.plot(x1,y1,label='background')
ax1.plot(x2,y2,label='signal')
ax1.axvline(0.5,color='r')
ax1.legend()

fpr = np.load('C:\\Users\\user\\Documents\\Fifth Year\\invisible-higgs\\src\\rnn_roc_results\\fpr.npy')
tpr = np.load('C:\\Users\\user\\Documents\\Fifth Year\\invisible-higgs\\src\\rnn_roc_results\\tpr.npy')
roc_auc = np.load('C:\\Users\\user\\Documents\\Fifth Year\\invisible-higgs\\src\\rnn_roc_results\\auc.npy')

ax2.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
ax2.plot([0, 1], [0, 1], 'k--')
    
ax2.set_xlim([0.0, 1.0])
ax2.set_ylim([0.0, 1.05])
    
ax2.legend(loc="lower right")
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
```
